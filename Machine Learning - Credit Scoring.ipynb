```python
x = dataset.drop(['default'], axis=1)
y = dataset['default']

SEED = 77
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=SEED)

base_train = x_train.shape[0]
base_test = x_test.shape[0]

print(f'A base de treino tem: {base_train} elementos, a base de teste tem: {base_test} elementos.')
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
print(model.score(x_train, y_train).round(2))
```

```python
previsoes = model.predict(x_test)
previsoes
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, previsoes)
accuracy
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix

confusion_matrix = plot_confusion_matrix(model, x_test, y_test, cmap='Blues')
confusion_matrix = confusion_matrix
```

(0,0) O modelo disse que é um BOM PAGADOR e realmente é um BOM PAGADOR

(0,1) O modelo disse que é um BOM PAGADOR porém  é um MAL PAGADOR

(1,0) O modelo disse que é um MAL PAGADOR porém é um BOM PAGADOR

(0,1) O modelo disse que é um MAL PAGADOR e realmente é um MAL PAGADOR

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, previsoes))
```

Precisão: Proporção de predições corretas de uma categoria em relação a todas as predições desta categoria.

Recall: Proporção de predições corretas de categoria alvo.

F1-Score: Média harmônica entre precisão e recall.

```python
from sklearn.metrics import roc_auc_score, roc_curve

prob_previsao = model.predict_proba(x_test)[:,1]

#tfp = taxa falsos positivos
#tvp = taxa verdadeiros positivos

tfp, tvp, limit = roc_curve(y_test, prob_previsao)
print('roc_auc', roc_auc_score(y_test, prob_previsao).round(2))

plt.subplots(1, figsize=(5,5))
plt.title('Curva ROC')
plt.plot(tfp, tvp)
plt.xlabel=('Especifidade')
plt.ylabel=('Sensibilidade')
plt.plot([0, 1], ls="--", c = 'red') #plotando linha guia pontilhada vermelha
plt.plot([0, 0], [1, 0], ls="--", c = 'green'), plt.plot([1, 1], ls="--", c = 'green') #plotando linha guia pontilhada verde
plt.show
```
